문제 : Write a program that prints the numbers from 1 to 100, each separated by a
newline.
If the number is a multiple of 3, it prints 'fizz' instead.
If the number is a multiple of 5, it prints 'buzz' instead.
If the number is both a multiple of 3 and a multiple of 5, it prints 'fizzbuzz' instead.

번역 : 1부터 100까지의 숫자를 출력하는 프로그램을 작성하십시오. 각각은 개행으로 구분됩니다.
숫자가 3의 배수이면 'fizz’를 출력합니다.
숫자가 5의 배수이면 'buzz’를 출력합니다.
숫자가 3과 5의 배수 모두인 경우 'fizzbuzz’를 출력합니다.


풀이 방법
A. 1~100까지 숫자를 다음 레이블로 출력하는 코드를 구성한다

```python
def fizzbuzz():
    for i in range(1,100):
        print(i)
```

```python
fizzbuzz()
```

B. 3의 배수일 때 문자"fizz"를 출력한다
 => IF 문을 사용한다

```python
def fizzbuzz():
    for i in range(1,100):
        if(i%3==0):
            print("fizz")
        else:
            print(i)
fizzbuzz()
```

C. 숫자가 5의 배수일 때 문자 "buzz"를 출력한다 => 다중 if 를 사용한다

```python
def fizzbuzz():
    for i in range(1,100):
        if(i%3==0):
            print("fizz")
        elif(i%5==0):
            print("buzz")
        else:
            print(i)
fizzbuzz()
```

D. 숫자가 3,5의 배수 모두인 경우 "fizzbuzz를 출력한다 => 다중 IF 조건에 AND 를 사용한다

```python
def fizzbuzz():
    for i in range(1,100):
        if(i%3==0 and i%5==0):
            print("fizzbuzz")
        elif(i%3==0):
            print("fizz")
        elif(i%5==0):
            print("buzz")
        else:
            print(i)
fizzbuzz()
```

E. 오류 수정 => Example 출력에서는 98, fizz, buzz 를 출력한다 그러나 조건은 1~100까지 출력한다 
=> 그러므로 98, fizz, buzz 의 출력을 위해서는 값을 1~101로 늘린다 (101)

```python
def fizzbuzz():
    for i in range(1,101):
        if(i%3==0 and i%5==0):
            print("fizzbuzz")
        elif(i%3==0):
            print("fizz")
        elif(i%5==0):
            print("buzz")
        else:
            print(i)
fizzbuzz()
```
